# Teile und Herrsche

## rekursive Suche

Aufsteigend sortierte Liste

Suche ein element

```python
a = [1, 3, 5, 7, 9, 11, 13, 15]


def binary_search(arr, li, re, key):
    if li > re:
        return -1

    mid = (li + re) // 2

    if arr[mid] == key:
        return mid
    elif key < arr[mid]:
        return binary_search(arr, li, mid - 1, key)
    else:
        return binary_search(arr, mid + 1, re, key)

binary_search(a, 0, len(a) - 1, 9)
```

## Klausur Binärdarstellung

Entwerfen Sie einen Algorithmus nach dem Teile-und-Herrsche-Prinzip zur Berechnung der Binärdarstellung einer ganzen Zahl n


```python
def binär(n):
    if n == 0:
        return
    binär(n//2)
    print(n%2, end="")

binär(13)
```

## Klausur Anzahl Vertauschungen in Liste

```python
arr = [2, 3, 8, 6, 1]

def mergeAndCount(arr, left, right):
    if left >= right:
        return 0

    mid = (left + right) // 2

    inv_count = mergeAndCount(arr, left, mid)
    inv_count += mergeAndCount(arr, mid + 1, right)

    inv_count += merge(arr, left, mid, right)

    return inv_count

def merge(arr, left, mid, right):
    temp = []
    i = left
    j = mid + 1
    inv_count = 0

    while i <= mid and j <= right:
        if arr[i] <= arr[j]:
            temp.append(arr[i])
            i += 1
        else:
            temp.append(arr[j])
            inv_count += (mid - i + 1) # alle anderen i sind auf jeden fall auch größer
            j += 1

    # Hänge die übrigen an die liste an
    while i <= mid:
        temp.append(arr[i])
        i += 1
    while j <= right:
        temp.append(arr[j])
        j += 1

    for k in range(len(temp)):
        arr[left + k] = temp[k]

    return inv_count

mergeAndCount(arr, 0, len(arr) - 1)
```

## Quicksort



```python
def quicksort(arr, left, right):
    if left >= right:
        return

    pivot_index = partition(arr, left, right)
    quicksort(arr, left, pivot_index - 1)
    quicksort(arr, pivot_index + 1, right)

def partition(arr, left, right):
    pivot = arr[right]      # könnte man auch random wählen für manchmal bessere Laufzeit
    i = left - 1

    for j in range(left, right):
        if arr[j] <= pivot:
            i += 1
            arr[i],  arr[j] = arr[j], arr[i]

    arr[i + 1], arr[right] = arr[right], arr[i + 1]
    return i + 1

arr = [8, 3, 1, 7, 0, 10, 2]
print("Unsortiert:", arr)
quicksort(arr, 0, len(arr) - 1)
print("Sortiert:  ", arr)
```
